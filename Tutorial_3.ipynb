```python
# 1. create a numpy array of size 2*2
import numpy as np
arr_2x2 = np.array([[1,2],[3,4]])
print(arr_2x2)
```

```python
# 2. create a random numpy array of size 3*4
arr_random_3x4 = np.random.rand(3,4)
print(arr_random_3x4)
```

```python
# 3. create a numpy array of ones and a numpy array of zeroes
arr_ones = np.ones((2,3))
arr_zeroes = np.zeros((2,3))
print(arr_ones)
print(arr_zeroes)
```

```python
# 4. create a random numpy array of size 6*3 and transpose it
arr_6x3 = np.random.rand(6,3)
arr_transposed = arr_6x3.T
print(arr_transposed)
```

```python
# 5. create two random numpy array of size n*p and p*m and then compute the dot product
import numpy as np

n=4
p=2
m=5

arr_nxp = np.random.rand(n,p)
arr_pxm = np.random.rand(p,m)
dot_product = np.dot(arr_nxp,arr_pxm)

print(arr_nxp)
print(arr_pxm)
print(dot_product)
```
